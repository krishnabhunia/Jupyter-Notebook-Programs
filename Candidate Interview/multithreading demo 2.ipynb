```python
import threading
import time
import os

def print_cube(n):
    time.sleep(2)
    print(f"The Cube of {n} is {n*n*n}")
    print(threading.currentThread())
    
def print_sq(n):
    print(f"The Square of {n} is {n*n}")
    print(threading.currentThread())
    
def isA(pid):
    for p in pid:
        print(p.is_alive())
    
    
if __name__ == "__main__":
    
    print(f"\nThe number of active threads are :{threading.activeCount()}")
#     print(dir(os))
    
    t1 = threading.Thread(target=print_cube,args=(10,))
    t2 = threading.Thread(target=print_sq,args=(10,))
    
    t3 = threading.Thread(target=print_cube,args=(10,))
    t4 = threading.Thread(target=print_sq,args=(10,))
    
    print(f"PID : {os.getpid()} \nPPID : {os.getppid()}")
    
    print(f"\nThe number of active threads are :{threading.activeCount()}")
    
    print(isA([t1,t2,t3,t4]))
    
    
    t1.start()
    t2.start()
    
    t3.start()
    t4.start()
    
    for titem in threading.enumerate():
        print(titem)
    
    print(f"\nThe number of active threads are :{threading.activeCount()}")
    
    t1.join()
    t2.join()
    
    t3.join()
    t4.join()
    
    print(f"\nThe number of active threads are :{threading.activeCount()}")
    
    print("\nDone")
    
```
